```python
%pip install numpy
%pip install matplot
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate sample data
np.random.seed(0)
x = np.arange(1, 11)
y = np.random.randint(1, 20, size=10)
data = np.random.randn(1000)
categories = ['A', 'B', 'C', 'D']
values = [15, 30, 45, 10]
```

```python
# Line Plot
plt.figure(figsize=(10, 6))
plt.subplot(2, 3, 1)
plt.plot(x, y, marker='o', linestyle='-', color='b')
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
# Scatter Plot
plt.subplot(2, 3, 2)
plt.scatter(x, y, color='r')
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
```

```python
# Bar Plot
plt.subplot(2, 3, 3)
plt.bar(categories, values, color='g')
plt.title('Bar Plot')
plt.xlabel('Category')
plt.ylabel('Value')
```

```python
# Histogram
plt.subplot(2, 3, 4)
plt.hist(data, bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
```

```python
# Box Plot
plt.subplot(2, 3, 5)
plt.boxplot(data, patch_artist=True)
plt.title('Box Plot')
plt.xlabel('Category')
plt.ylabel('Value')
```

```python
# Pie Chart
plt.subplot(2, 3, 6)
plt.pie(values, labels=categories, colors=['gold', 'lightblue', 'lightgreen', 'lightcoral'], autopct='%1.1f%%')
plt.title('Pie Chart')

plt.tight_layout()
plt.show()
```

```python
# Area Plot
plt.figure(figsize=(10, 6))
plt.fill_between(x, y, color='orange', alpha=0.5)
plt.title('Area Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```
